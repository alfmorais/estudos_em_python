```
import pandas as pd

alunos = ['Alfredo', 'Denise', 'Joaquim', 'Vilma', 'Viviane', 'Derek', 'Ricardo',
          'Joana', 'Renata', 'Bruna', 'Luciana', 'Joilson', 'Kyra', 'Ryan']

nota_1 = [8.0, 9.0, 7.0, 7.5, 8.0, 6.0, 5.6, 8.0, 9.0, 7.0, 7.5, 8.0, 6.0, 5.6]

nota_2 = [9.0, 7.0, 8.0, 6.5, 10.0, 6.5, 7.6, 8.5, 9.5, 7.2, 7.5, 8.7, 6.8, 6.9]

media = []

for nota1, nota2, in zip(nota_1, nota_2):
  media_final = round(((nota1 + nota2) / 2), 2)
  media.append(media_final)

data_alunos = pd.DataFrame({
    'Alunos': alunos,
    'Nota 1': nota_1,
    'Nota 2': nota_2,
    'Media Final': media,
})

data_alunos
```
