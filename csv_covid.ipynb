```
import csv 

caminho = '/content/WHO COVID-19 global table data February 12th 2021 at 9.46.34 AM.csv'
```

```
with open(caminho, newline='') as cvsfile:
  arquivo = csv.reader(cvsfile, delimiter=' ', quotechar='|')
  lista_arquivo = []
  for linha in arquivo:
    print(' '.join(linha) + '\n')
    lista_arquivo.append(' '.join(linha))

print(lista_arquivo)
```

```
contador = []
for linha in lista_arquivo:
  valor = linha.count(',')
  valor = int(valor)
  contador.append(valor)
```

```
print(contador)
```
