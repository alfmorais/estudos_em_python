```
import random
import pandas as pd

# Lista de numeros possiveis 
lista_numeros = list(range(1, 26, 1))

# contador do loop while
contador = 0

# lista de palpites
lista_palpites = []

while True:
  # Usando a função random.choices para selecionar os números
  palpites = random.choices(lista_numeros, k=15)
  
  # Convertendo a lista em set
  palpites = set(palpites)
  
  # verificando se o palpites tem 15 digitos
  if len(palpites) == 15:
    print(palpites)
    contador += 1
    palpites = list(palpites)
    lista_palpites.append(palpites)

    # verificando se o palpites foi impresso 10 vezes e saindo do loop while
    if contador == 10:
      break
```

```
# Elaborando o DataFrame
df_palpites = pd.DataFrame({
    'Jogo 1': lista_palpites[0],
    'Jogo 2': lista_palpites[1],
    'Jogo 3': lista_palpites[2],
    'Jogo 4': lista_palpites[3],
    'Jogo 5': lista_palpites[4],
    'Jogo 6': lista_palpites[5],
    'Jogo 7': lista_palpites[6],
    'Jogo 8': lista_palpites[7],
    'Jogo 9': lista_palpites[8],
    'Jogo 10': lista_palpites[9]
})

# Imprimindo o dataframe
df_palpites
```

```
# Fazendo o arquivo palpites
with open('palpites.txt', 'w') as file:
  file.write(
      df_palpites.to_string()
  )
```
